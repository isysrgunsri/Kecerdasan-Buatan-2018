```python
X = data.iloc[:,0:5].values
Y = data.iloc[:,5].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.fit_transform(x_test)
```

BACK PROPAGATION

pada bagian ini Hidden Layers, input dan ouput layer ditambahkan, dengan kedua input layer menggunakan aktivasi ReLU
sedangkan output layer menggunakan aktivasi sigmoid.

dari input menuju ke hidden layer 1, hidden layer 1 ke hidden layer 2, dan hidden layer 2 menuju ke outputdisebut Forward Pass,
karena aktivasinya menggunakan ReLU maka menggunakan f(x)=max(0,x)



lalu untuk aktivasi sigmoid akan diturunkan menjadi : 


selanjutnya adalah backward pass yang akan mengalirkan loss keseluruh node - node sehingga didapatkan gradient loss
1. mencari turunan parsial:
    
2. mencari gradien Oout terhadap Oin karena aktivasi sigmoid maka menggunakan turunan yang telah disebutkan sebelumnya
3. mecari gradient Oin terhadap Wk1o...Wkno dan bias (bo)
4. lalu menggunakan chain rule


Setelah backward pass akan dilakukan SGD Update yaitu hidden layer 2 ke hidden layer 1

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential()

# Hidden layers and input layer are  added.
classifier.add(Dense(3, kernel_initializer="uniform", activation = 'relu', input_dim = 5))

classifier.add(Dense(3, kernel_initializer="uniform", activation = 'relu'))

# Output layer
classifier.add(Dense(1, kernel_initializer="uniform", activation = 'sigmoid'))

classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

classifier.fit(X_train, y_train, epochs = 50)

# Prediction
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_pred = (y_pred > 0.5)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```
