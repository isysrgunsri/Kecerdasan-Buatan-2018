```python
# Memasukkan Libary 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Memasukan dataset
data = pd.read_csv('data.csv')
del data['Unnamed: 32']
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
X = data.iloc[:, 2:].values
y = data.iloc[:, 1].values

# Encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
y = labelencoder_X_1.fit_transform(y)

# Membagi data set menjadi train data dan test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)

# Memberikan fitur scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
# Menginisiasi ANN
classifier = Sequential()
```

```python
# menambahkan Hidden layer
classifier.add(Dense(16,kernel_initializer='uniform', activation='relu', input_dim=30))
# menambahkan overpitting
classifier.add(Dropout(0.1))
```

```python
# Menambahkan Hidden Layer Kedua
classifier.add(Dense(units=16, kernel_initializer='uniform', activation='relu'))
# Menambahkan Overpitting
classifier.add(Dropout(0.1))
```

```python
# Menambahkan Output
classifier.add(Dense(units=1,kernel_initializer='uniform', activation='sigmoid'))
```

```python
# Mengcompile data ANN
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Menyamakan model ANN dengan data test
classifier.fit(X_train, y_train, batch_size=100, epochs=150)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Membuat confusion martix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print("Our accuracy is {}%".format(((cm[0][0] + cm[1][1])/57)*100))
```

```python
sns.heatmap(cm,annot=True)
plt.savefig('h.png')
```
