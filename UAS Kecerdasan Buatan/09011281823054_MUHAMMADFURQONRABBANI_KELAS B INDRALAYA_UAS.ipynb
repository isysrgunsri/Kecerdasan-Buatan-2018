Ada teknik lainnya yang disebut dengan One-Hot Encoding

One-Hot Encoding adalah teknik yang merubah setiap nilai di dalam kolom menjadi kolom baru dan mengisinya dengan nilai biner yaitu 0 dan 1.
Pada tabel Gender sudah menjadi 0 dan 1 jadi tidak perlu kita lakukan proses OneHotEncoding


```python
#One hot encoding. 
#untuk melakukan one hot encoding kita menggunakan to_categorical dari package keras
from keras.utils import to_categorical
encoded = pd.DataFrame(to_categorical(x.iloc[:, 1]))
print(encoded)
#selanjutnya kita gabungkan data ini ke tabel x
#no need to encode Gender, as there are only two categories
x = pd.concat([encoded, x], axis = 1)
x.head()
```

karena kolom geografi telah diencode menjadi kolom 0 1 2 maka kita akan hapus. kemudian kita hapus salah satu kolom
hasil dari one hot encoder karena dengan dua kolom saja sudah dapat kita bedakan.

```python
#Dropping the existing "geography" category, and one of the onehotcoded columns.

x = x.drop(['Geography', 0], axis = 1)
x.head(10)
```

Dapat kita lihat disini jika nilai pada kolom 1 dan 2

0 1 merepresentasikan geografi -> Germany

Terakhir, untuk praproses data sebelum membangun model, kita bagi data dari X menjadi 2 yaitu untuk training model dan sisanya untuk mengecek pemakaian model. rasio pembagiannya untuk ini adalah 80 % untuk di training dan 20% untuk di test. Selanjutnya input X baik train maupun tes kita standarisasi hal ini ditujukan agar pembangunan model lebih efektif.

```python
#train and test set split, and feature scaling

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### 2. Build the model using Keras
Disini kita membuat modelnya
dengan model sequential
dense untuk menambah layer yang full connected


```python
import keras
from keras.models import Sequential
from keras.layers import Dense #to add layers

#there is no rule on how many nodes each hidden layer should have
classifier = Sequential()
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 11))
#init --> initialize weights according to uniform distribution
#input_dim is required for the first hidden layer, as it is the first starting point. --> number of nodes.
#output_dim --> number of nodes of the hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))
#input_dim --> remove it as it already knows what to expect.

#the output layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
#output_dim should be 1, as output is binary outcome, and activation should be 'sigmoid'
#If dependent variables have more than two categories, use activation = 'softmax'

#compile the model --> backpropagation -> gradient descent
classifier.compile(optimizer = 'adam', loss = "binary_crossentropy", metrics = ['accuracy'])
#optimizer = algorithm to find the optimal set of weights in ANN
#loss = functions that should be optimized. if more than two categories, use "categorical_crossentropy"
#metrics = criterion used to calculate the performance of the model.
```

Selanjutnya kita jalankan model dengan batch size 10 dan epoch 20 dengan inputan X, Train dan, Y_train

```python
classifier.fit(X_train, Y_train, batch_size = 10, nb_epoch = 20)
#batch_size = the number of observations after which you want to update the weights
#           batch size and epochs should be tuned through experiments.
#epoch = going through the whole dataset
```

Selanjutnya untuk mengecek akurasi, kita buat prediksi sendiri dengan model kita dan bandingkan dengan output  yang sebenarnya dengan confusion matrix.
apabila output dari prediksi lebih dari 0.5 true atau benar dibawah itu false atau salah. kemudian dengan 

```python
#predicting the results

y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5) #to classify each probability into True or False

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
print (cm, '\n\n', y_pred[:5, :])
```
