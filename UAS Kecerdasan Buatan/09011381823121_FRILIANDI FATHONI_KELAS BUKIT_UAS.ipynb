# Klasifikasi menggunakan Haberman's Survival Data Set

Ini adalah implementasi algoritma K-Nearest Neighbours menggunakan Python.

**K-Nearest Neighbor** atau **KNN** adalah salah satu algoritma yang digunakan untuk melakukan klasifikasi terhadap objek berdasarkan dari data pembelajaran (data training) yang jaraknya paling dekat dengan objek tersebut. Tujuan dari algoritma KNN adalah mengklasifikasikan objek baru berdasarkan atribut dan sampel-sampel dari data training.


**Menambahkan module math**

```python
import math
```

```python
data = []
```

##### Impor data dan tambahkan ke daftar

[age_of_the_patient, year_of_operation, number_of_nodes_detected, survival_status]



```python
with open('dataset.data', 'r') as f:
    for line in f.readlines():
        atributes = line.strip('\n').split(',')
        data.append([int(x) for x in atributes])
```

##### Fungsi pembantu untuk membantu visualisasi
Juga mengembalikan informasi kunci dari kumpulan data

```python
def info_dataset(data, verbose=True):
    label1, label2 = 0, 0
    data_size = len(data)
    for datum in data:
        if datum[-1] == 1:
            label1 += 1
        else:
            label2 += 1
    if verbose:
        print('Total of samples: %d' % data_size)
        print('Total label 1: %d' % label1)
        print('Total label 2: %d' % label2)
    return [len(data), label1, label2]
```

```python
info_dataset(data)
```

##### Tentukan kereta / persentase total

```python
p = 0.6
_, label1, label2 = info_dataset(data,False)
```

##### Memisahkan kumpulan data menjadi set kereta dan set pengujian

```python
train_set, test_set = [], []
max_label1, max_label2 = int(p * label1), int(p * label2)
total_label1, total_label2 = 0, 0
for sample in data:
    if (total_label1 + total_label2) < (max_label1 + max_label2):
        train_set.append(sample)
        if sample[-1] == 1 and total_label1 < max_label1:
            total_label1 += 1
        else:
            total_label2 += 1
    else:
        test_set.append(sample)
```

```python
def euclidian_dist(p1, p2):
    dim, sum_ = len(p1), 0
    for index in range(dim - 1):
        sum_ += math.pow(p1[index] - p2[index], 2)
    return math.sqrt(sum_)
```

##### Menghitung jarak antara sampel tertentu dan sampel lainnya dalam set kereta
Memberi jarak ke kamus, mengurutkannya, dan mendapatkan K tetangga terdekat;
Kemudian itu menghitung label dan mengembalikannya.

```python
def knn(train_set, new_sample, K):
    dists, train_size = {}, len(train_set)
    
    for i in range(train_size):
        d = euclidian_dist(train_set[i], new_sample)
        dists[i] = d
    
    k_neighbors = sorted(dists, key=dists.get)[:K]
    
    qty_label1, qty_label2 = 0, 0
    for index in k_neighbors:
        if train_set[index][-1] == 1:
            qty_label1 += 1
        else:
            qty_label2 += 1
            
    if qty_label1 > qty_label2:
        return 1
    else:
        return 2
```

##### Example

```python
print(test_set[0])
print(knn(train_set, test_set[0], 12))
```

##### Menghitung prediksi yang benar dari set pengujian dengan K tertentu

```python
correct, K = 0, 15
for sample in test_set:
    label = knn(train_set, sample, K)
    if sample[-1] == label:
        correct += 1
```

```python
print("Train set size: %d" % len(train_set))
print("Test set size: %d" % len(test_set))
print("Correct predicitons: %d" % correct)
print("Accuracy: %.2f%%" % (100 * correct / len(train_set)))
```
